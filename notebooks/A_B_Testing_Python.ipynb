### A/B Test Process

A complete A/B test typically includes:

1. Analyze current state and form hypotheses: Identify the highest-priority improvements and propose optimizations.

2. Define metrics: Set primary metrics to compare variants and secondary metrics to capture other effects.

3. Design and develop: Build a prototype of the new variant and implement it.

4. Determine test duration: Decide how long the experiment will run.

5. Define allocation: Set traffic split and other assignment details for each variant.

6. Collect and analyze data: Gather experiment data and assess validity and effect.

7. Draw conclusions: (a) Ship the new version; (b) Adjust allocation and continue testing; or (c) Iterate on the design and return to step 1.

```python
import gdown
import pandas as pd
import numpy as np
import io

def load_from_drive(file_id, file_type, info=True):
    # Build download URL
    download_url = f'https://drive.google.com/uc?id={file_id}'

    # Set output filename
    output = f'downloaded_file.{file_type}'

    # Download file
    gdown.download(download_url, output, quiet=False)

    # Read by file extension
    if file_type == 'csv':
        read = pd.read_csv(output)  # Read CSV
    elif file_type == 'xlsx':
        read = pd.read_excel(output)  # Read Excel
    else:
        raise ValueError("Unsupported file type. Please use 'csv' or 'xlsx'.")

    # Print data info (optional)
    if info:
        if len(read) > 0:
            print("# Data loaded!")
            print("# ------------------------------------", "\n")

            # Dimensions
            print("# Dimensions -------------------------")
            print("Rows:", read.shape[0], "Columns:", read.shape[1], "\n")

            # Data types
            print("# Data types -----------------------------")
            if len(read.select_dtypes("object").columns) > 0:
                print("Object columns:", "\n", "Count:",
                      len(read.select_dtypes("object").columns), "\n",
                      read.select_dtypes("object").columns.tolist(), "\n")

            if len(read.select_dtypes("int64").columns) > 0:
                print("Integer columns:", "\n", "Count:",
                      len(read.select_dtypes("int64").columns), "\n",
                      read.select_dtypes("int64").columns.tolist(), "\n")

            if len(read.select_dtypes("float64").columns) > 0:
                print("Float columns:", "\n", "Count:",
                      len(read.select_dtypes("float64").columns), "\n",
                      read.select_dtypes("float64").columns.tolist(), "\n")

            if len(read.select_dtypes("bool").columns) > 0:
                print("Boolean columns:", "\n", "Count:",
                      len(read.select_dtypes("bool").columns), "\n",
                      read.select_dtypes("bool").columns.tolist(), "\n")

            # Missing values
            print("# Missing values ---------------------")
            print("Any missing values? \n ", np.where(read.isnull().values.any() == False,
                                            "No missing values!", "Data contains missing values!"), "\n")

            # Memory usage
            buf = io.StringIO()
            read.info(buf=buf)
            info = buf.getvalue().split('\n')[-2].split(":")[1].strip()
            print("# Memory usage ---------------------- \n", info)

        else:
            print("# Data not loaded!")

    return read

# Example usage
file_id = '1IkN-fylT9ZYxZJhzbgpAmJ751b6lrPy9'  # Replace with your file ID
file_type = 'csv'  # 'csv' or 'xlsx'
ab = load_from_drive(file_id, file_type)
ab.head()
```

#### Experiment steps

Assumptions:

- Check normality
- If normal, check homogeneity of variance

Steps:

- Split and define control vs treatment
- Shapiro test for normality
- If parametric assumptions hold, Levene test for equality of variance
- If parametric and equal variance: t-test
- If parametric but unequal variance: Welch test
- If non-parametric: Mann-Whitney U test

**Flowchart summary:** Start → Are groups defined? (No → define control and test groups). Apply Shapiro–Wilk normality test. If normality fails → use Mann–Whitney U. If normality holds → Levene's test for homogeneity of variance. If homogeneous → use T-test; if not → use Welch's test.\n

#### Notes

<font color = red>Notes:</font>
1. Test duration: The test should not be too short, or participants will mostly be power users.

2. Traffic allocation (sampling): Ensure simultaneity, homogeneity, uniqueness, and uniformity.

① Simultaneity: Allocation and test execution should be simultaneous.

② Homogeneity: Split user groups should be similar on all dimensions (device, OS, language, demographics, tenure, etc.). Each A/B test can target a specific segment.

How to check homogeneity? Use AAB: run two A groups and one B. If A1 vs A2 shows no significant difference, allocation is likely homogeneous.

③ Uniqueness: Users must not be counted in more than one group.

④ Uniformity: Traffic should be evenly split (e.g. via hashing). Often handled by A/B testing platforms. One approach: hash (user_id + salt) to a value in [0,1] and compare to a threshold to assign groups.

#### Why check for normality?

Many methods (e.g. t-test) are more accurate when data are normal. Normality checks help confirm that the chosen test is appropriate and avoid wrong conclusions.

If data are normal, parametric methods (e.g. t-test) can estimate differences in means more accurately. If not, parametric methods may be biased or invalid.

In A/B testing we use normality tests to validate our choice of method. If data are not normal, we can use non-parametric methods (e.g. Mann-Whitney U) that do not assume a distribution.



#### What if data are not normal?

Use non-parametric tests, which do not assume normality and are more flexible.

Common non-parametric methods:

1. Mann-Whitney U: compare medians of two independent samples (most common here).
2. Wilcoxon signed-rank: compare two related samples.
3. Kruskal-Wallis: compare multiple independent samples.
4. Friedman: compare multiple related samples.

#### Why test for equality of variance?

To ensure that the independent two-sample t-test (and similar parametric tests) are valid. Homoscedasticity means the groups have similar variance.

If data are normal and variances are equal, use the two-sample t-test (ttest_ind). If not normal or unequal variance, use Mann-Whitney U (mannwhitneyu).

#### Non-parametric tests


|            | Non-parametric                                                | Parametric                                                    |
| ---------- | -------------------------------------------------------------- | ------------------------------------------------------------- |
| Distribution | Unknown or non-normal                                       | Normal                                                        |
| Equal variance | Not required                                              | Required for two+ groups                                      |
| Variable type | Ordinal / non-normal continuous                            | Interval/ratio                                                |
| Power       | Generally lower                                              | Generally higher when assumptions hold                        |
| Pros        | Fewer assumptions; widely applicable.                        | Higher power when assumptions hold.                          |
| Cons        | Lower power; less precise when data are normal.               | Narrow applicability; requires normality.                   |

When to use non-parametric methods:
1. Sparse metrics;
2. Extreme outliers;
3. Clearly non-normal distribution.

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from scipy.stats import shapiro, ttest_ind, mannwhitneyu, levene
import statsmodels.stats.api as sms

def AB_Test(dataframe, group, target, alpha=0.05, visualize=True):
    '''
    Run A/B test on numeric target.
    '''
    # Split A/B groups
    groupA = dataframe[dataframe[group] == "A"][target]
    groupB = dataframe[dataframe[group] == "B"][target]

    # Normality test
    ntA = shapiro(groupA)[1] < alpha
    ntB = shapiro(groupB)[1] < alpha

    # Homogeneity of variance
    leveneTest = levene(groupA, groupB)[1] < alpha

    if (ntA == False) & (ntB == False):  # normal
        # Equal variance
        if leveneTest == False:
            ttest, p_value = ttest_ind(groupA, groupB, equal_var=True)
            test_type = "Parametric (Equal Variance)"
        else:
            ttest, p_value = ttest_ind(groupA, groupB, equal_var=False)
            test_type = "Parametric (Unequal Variance)"
    else:
        # Non-parametric test
        ttest, p_value = mannwhitneyu(groupA, groupB, alternative='two-sided')
        test_type = "Non-Parametric"

    # Mean difference
    mean_diff = groupB.mean() - groupA.mean()

    # Effect size
    nobs_A = len(groupA)
    nobs_B = len(groupB)
    pooled_std = np.sqrt(((nobs_A-1)*groupA.var() + (nobs_B-1)*groupB.var()) / (nobs_A + nobs_B - 2))
    effect_size = mean_diff / pooled_std

    # Observed power
    observed_power = sms.tt_ind_solve_power(effect_size=effect_size, nobs1=nobs_A, alpha=alpha, alternative='two-sided')

    # Result
    ab_hypothesis = p_value < alpha
    comment = "No significant difference between A/B groups!" if not ab_hypothesis else "Significant difference between A/B groups!"

    result = pd.DataFrame({
        "Test Type": [test_type],
        "target": [target],
        "Homogeneity": ["Yes" if not leveneTest else "No"],
        "AB Hypothesis": ["Reject H0" if ab_hypothesis else "Fail to Reject H0"],
        "p-value": [p_value],
        "Mean Difference": [mean_diff],
        "Effect Size": [effect_size],
        "Observed Power": [observed_power],
        "Comment": [comment],
        "t-stats": [ttest]
    })

    # Print hypotheses
    print("# A/B test hypotheses")
    print("H0: A == B")
    print("H1: A != B", "\n")

    # Normality and homogeneity
    print(f"Normality satisfied?\nGroup A: {not ntA}, Group B: {not ntB}\n")
    print(f"Homogeneity of variance: {not leveneTest}", "\n")

    # Visualize
    if visualize:
        plt.figure(figsize=(6, 4))
        sns.histplot(data=dataframe, x=target, hue=group, kde=True)
        plt.axvline(groupA.mean(), color='blue', linestyle='dashed', linewidth=2, label=f'Mean A: {groupA.mean():.2f}')
        plt.axvline(groupB.mean(), color='orange', linestyle='dashed', linewidth=2, label=f'Mean B: {groupB.mean():.2f}')
        plt.title('Distribution of Groups')
        plt.legend()
        plt.show()

    return result

# Run A/B test
result = AB_Test(dataframe=ab, group="version", target="sum_gamerounds")

result
```

### 7. A/B test (binary 0-1 metrics)

Metrics like retention = retained users / total users, or cold-start pass rate = plans reaching 20 converts in 3 days / total new plans, are 0-1 variables; we then apply a two-sample test for proportions (or t-test on the binary outcome).


Proportion metrics (e.g. UV conversion, next-day retention) are usually “count under condition / total count”; we use a two-sample test for proportions. For a single user’s retention, the outcome is binary: either they retained or not.

The number of successes in n trials follows a binomial distribution.

When n is large, the sampling distribution of the sample proportion is approximately normal, so we can use a z-test for the equality of two proportions.


$$
z=\frac{{p_1} - {p_2}}{\sqrt{({p}(1-{p})(\frac{1}{n_1}+\frac{1}{n_2})}}
$$

- $ p_1 $ and $ p_2 $ are the two sample proportions (e.g. retention in group A and B).
- $ n_1 $ and $ n_2 $ are the sample sizes.
- $ p $ is the pooled proportion: $ p = \frac{n_1 p_1 + n_2 p_2}{n_1 + n_2} $ (weighted average).
- $ 1 - p $ is the complementary proportion.

#### Bootstrapping

**Why Bootstrapping is useful in A/B testing:**

1. **Estimating population parameters:** No distribution assumption is needed; we can estimate means, medians, standard deviations from the sample. Bootstrapping improves accuracy by resampling with replacement to approximate the sampling distribution of statistics.

2. **Assessing reliability of inference:** Resampling quantifies sampling error. In A/B tests we can see how statistics (e.g. retention rate differences) vary across resamples and evaluate how reliable our conclusions are. We can compute confidence intervals (e.g. for the difference in retention between groups); narrower intervals mean more precise estimates.

3. **Hypothesis testing:** When sample sizes are small or distributions are non-normal, bootstrap offers a non-parametric alternative. We can compare statistics between groups across resamples to test for statistically significant differences.


> **Definiton of Method:** We'll use bootstrapping for A/B testing. The basic idea of bootstrapping is that inference about a population from sample data (sample → population) can be modelled by resampling the sample data and performing inference about a sample from resampled data (resampled → sample). As the population is unknown, the true error in a sample statistic against its population value is unknown. In bootstrap-resamples, the 'population' is in fact the sample, and this is known; hence the quality of inference of the 'true' sample from resampled data (resampled → sample) is measurable.

1. **Sample data**: We start with the observed data (e.g. retention for the two groups in Cookie Cats).
2. **Resampling**: We draw many bootstrap samples by sampling with replacement; each has the same size as the original.
3. **Statistic**: For each bootstrap sample we compute the statistic of interest (e.g. difference in retention, Delta).
4. **Inference**: We use the distribution of the bootstrap statistic to assess uncertainty (e.g. standard error, confidence interval).

Its advantage is that it does not assume a population distribution; it uses the sample data directly to infer properties of the statistic.

Steps:

1. **Create 500 bootstrap samples**: Draw 500 samples with replacement from the original data, each of the same size.
2. **Compute 1-day and 7-day retention for A/B**: For each bootstrap sample, compute retention_1 and retention_7 for gate_30 (A) and gate_40 (B).
3. **Plot bootstrap distributions**: For 1-day and 7-day retention, plot the bootstrap distribution for each group.
4. **Compute retention difference**: For each bootstrap sample, compute the difference in retention between B and A.
5. **Probability that gate 30 is higher**: Estimate the probability that retention is higher under gate 30 than gate 40.
6. **Conclusion and recommendation**: Based on the analysis, recommend whether to keep the new setting or roll back.

```python
# Lists to store bootstrap results
boot_1d_diffs = []
boot_7d_diffs = []

# Number of bootstrap samples
n_bootstraps = 500  # can use 5000

# Resample
for i in tqdm(range(n_bootstraps)):
    boot_sample = df.sample(frac=1, replace=True)
    retention_1_A = boot_sample[boot_sample['version'] == 'A']['retention_1'].mean()
    retention_1_B = boot_sample[boot_sample['version'] == 'B']['retention_1'].mean()
    retention_7_A = boot_sample[boot_sample['version'] == 'A']['retention_7'].mean()
    retention_7_B = boot_sample[boot_sample['version'] == 'B']['retention_7'].mean()

    boot_1d_diffs.append(retention_1_B - retention_1_A)
    boot_7d_diffs.append(retention_7_B - retention_7_A)

# Convert to DataFrame
boot_1d_diffs = pd.Series(boot_1d_diffs)
boot_7d_diffs = pd.Series(boot_7d_diffs)

# 95% confidence interval
alpha = 0.05
ci_lower_1d = boot_1d_diffs.quantile(alpha/2)
ci_upper_1d = boot_1d_diffs.quantile(1 - alpha/2)
ci_lower_7d = boot_7d_diffs.quantile(alpha/2)
ci_upper_7d = boot_7d_diffs.quantile(1 - alpha/2)

# Print results
print(f"Retention 1-day: 95% CI = [{ci_lower_1d}, {ci_upper_1d}]")
print(f"Retention 7-day: 95% CI = [{ci_lower_7d}, {ci_upper_7d}]")

# Bootstrap distribution and CI
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(14, 6))
sns.histplot(boot_1d_diffs, kde=True, ax=ax1)
ax1.axvline(ci_lower_1d, color='r', linestyle='--')
ax1.axvline(ci_upper_1d, color='r', linestyle='--')
ax1.set_title('Bootstrapped Differences: 1-day Retention')
ax1.set_xlabel('Difference in Retention Rate')
ax1.set_ylabel('Frequency')

sns.histplot(boot_7d_diffs, kde=True, ax=ax2)
ax2.axvline(ci_lower_7d, color='r', linestyle='--')
ax2.axvline(ci_upper_7d, color='r', linestyle='--')
ax2.set_title('Bootstrapped Differences: 7-day Retention')
ax2.set_xlabel('Difference in Retention Rate')
ax2.set_ylabel('Frequency')

plt.tight_layout()
plt.show()

# Significance
if ci_lower_1d > 0 or ci_upper_1d < 0:
    print("Retention 1-day difference is significant.")
else:
    print("Retention 1-day difference is not significant.")

if ci_lower_7d > 0 or ci_upper_7d < 0:
    print("Retention 7-day difference is significant.")
else:
    print("Retention 7-day difference is not significant.")
```
